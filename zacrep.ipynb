```python
import pandas as pd
from IPython.display import display
import re
```

```python
ratings_movies=pd.read_csv('data\\ratings_movies.csv')
display(ratings_movies)
```

```python
def get_year_release(arg):
    candidates = re.findall(r'\(\d{4}\)',arg)
    if len(candidates)>0:
        year = candidates[0].replace('(','')
        year = year.replace(')','')
        return int(year)
    else:
        return None
```

```python
ratings_movies['year_release']=ratings_movies['title'].apply(get_year_release)
ratings_movies['year_release'].isna().sum()
```

```python
new_df = ratings_movies[ratings_movies['year_release']==2010].groupby('genres')['rating'].mean().nsmallest()
display(new_df)
```

```python
genres_unique =  ratings_movies.groupby('userId')['genres'].nunique().nlargest()
display(genres_unique)
```

```python
ratings_movies.groupby('userId')['rating'].agg(['count','mean']).sort_values(by ='count').sort_values(by = 'mean', ascending=True)
```
